# Python Special Methods (Dunder Methods)

Special methods in Python are easily recognizable because they begin and end with double underscores (e.g., `__init__`). They allow your custom objects to hook into Python's built-in operators and functions, enabling powerful object-oriented behaviors like operator overloading, custom string representations, and context management.

---

## 1. Initialization and Representation

These methods handle how an object is created and how it displays itself textually.

```python
class Book:
    def __new__(cls, *args, **kwargs):
        print("1. __new__: Allocating memory for the object.")
        return super().__new__(cls)

    def __init__(self, title, author, pages):
        print("2. __init__: Initializing the object's attributes.")
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        # User-friendly string representation (used by print() and str())
        return f"'{self.title}' by {self.author}"

    def __repr__(self):
        # Developer-friendly string representation (used in shell, debugging)
        return f"Book({self.title!r}, {self.author!r}, {self.pages})"

# Testing the methods
book = Book("The Hobbit", "J.R.R. Tolkien", 310)
print("str():", str(book))   # Triggers __str__
print("repr():", repr(book)) # Triggers __repr__
```

## 2. Mathematical Operator Overloading

These methods allow you to use standard math operators (`+`, `-`, `*`, `/`) with your custom classes.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        # Overloads the + operator
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        # Overloads the - operator
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        # Overloads the * operator (Vector * scalar)
        return Vector(self.x * scalar, self.y * scalar)

    def __truediv__(self, scalar):
        # Overloads the / operator
        return Vector(self.x / scalar, self.y / scalar)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

# Testing math operators
v1 = Vector(2, 4)
v2 = Vector(1, 3)

print("Addition:", v1 + v2)
print("Subtraction:", v1 - v2)
print("Multiplication:", v1 * 3)
print("Division:", v1 / 2)
```

## 3. Comparison Operators

These methods allow objects to be compared using `>`, `<`, `==`, etc. Implementing these also allows your objects to be sorted automatically using Python's `sorted()` function.

```python
class Box:
    def __init__(self, volume):
        self.volume = volume
    
    def __eq__(self, other):
        # Overloads ==
        return self.volume == other.volume
    
    def __lt__(self, other):
        # Overloads <
        return self.volume < other.volume
    
    def __le__(self, other):
        # Overloads <=
        return self.volume <= other.volume

# Testing comparisons
box1 = Box(20)
box2 = Box(50)

print("Is box1 == box2?", box1 == box2)
print("Is box1 < box2?", box1 < box2)
print("Is box1 <= box2?", box1 <= box2)
```

## 4. Emulating Container Types (Collections)

These methods make your custom objects behave like built-in lists, dictionaries, or sets.

```python
class CustomPlaylist:
    def __init__(self, songs):
        self.songs = songs  # Expects a list
        
    def __len__(self):
        # Overloads len()
        return len(self.songs)
        
    def __getitem__(self, index):
        # Overloads indexing syntax: playlist[index]
        return self.songs[index]
        
    def __setitem__(self, index, value):
        # Overloads index assignment: playlist[index] = value
        self.songs[index] = value
        
    def __contains__(self, item):
        # Overloads the 'in' keyword
        return item in self.songs

# Testing collection emulation
my_playlist = CustomPlaylist(["Song A", "Song B", "Song C"])

print("Length of playlist:", len(my_playlist))
print("First song:", my_playlist[0])

my_playlist[1] = "Song B (Remix)"
print("Updated playlist:", my_playlist.songs)
print("Is 'Song A' in playlist?", "Song A" in my_playlist)
```

## 5. Context Managers and Callables

These methods handle resource management (`with` blocks) and make objects behave like executable functions.
